```python
import re
import spacy

nlp = spacy.load("fr_core_news_sm")
stopwords_fr = nlp.Defaults.stop_words

def preprocess_text(text):
    text = re.sub(r"[^a-zA-ZÀ-ÿ0-9\s]", "", text)
    text = re.sub(r"\b(d'|l')([a-zA-ZÀ-ÿ]+)", r"\2", text)
    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc if token.lemma_ not in stopwords_fr and token.is_alpha]
    return tokens

tokens_filtrés = preprocess_text(conf)
print(tokens_filtrés)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_wordcloud(tokens):
    wordcloud = WordCloud(width=800, height=400, background_color=None, mode="RGBA", colormap="cool", max_words=100).generate(" ".join(tokens))
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()

tokens_filtrés = preprocess_text(conf)
generate_wordcloud(tokens_filtrés)
```

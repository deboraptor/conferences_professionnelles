```python
import re
import spacy
import collections
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import STOPWORDS

nlp = spacy.load("fr_core_news_sm")
stopwords_fr = nlp.Defaults.stop_words.union(STOPWORDS)

def preprocess_text(text):
    """ Nettoie et tokenize le texte en enlevant les stopwords et en lemmatisant les mots """
    text = re.sub(r"\b(d'|l')([a-zA-Z√Ä-√ø]+)", r"\2", text)
    doc = nlp(text)
    tokens = [token.lemma_.lower() for token in doc if token.lemma_ not in stopwords_fr and token.is_alpha]
    return tokens

def process_text(conference_text):
    """ Traite un texte de conf√©rence et renvoie un DataFrame des mots les plus fr√©quents """
    words = preprocess_text(conference_text)
    word_counts = collections.Counter(words)
    df = pd.DataFrame(word_counts.items(), columns=["Word", "Frequency"]).sort_values(by="Frequency", ascending=False)
    return df

for i, conf in enumerate(confs):
    word_frequencies = process_text(conf)
    display(word_frequencies.head(10))
    plt.figure(figsize=(12, 6))
    plt.bar(word_frequencies["Word"][:10], word_frequencies["Frequency"][:10], color="hotpink")
    plt.xticks(rotation=45)
    plt.show()
```
